###Impute missing numerical values

```python
inputs_df[input_cols].isna().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
inputs_df = imputer.fit_transform(inputs_df)
```

```python
inputs_df = pd.DataFrame(inputs_df, columns = input_cols)
inputs_df
```

###Scale numeric values
Let's scale the numeric values to a range of 0 to 1; Making it easier for our 

*   Let's scale the numeric values to a range of 0 to 1; Making it easier for our model to train.
*   In our case, all of our input columns are numeric values





```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(inputs_df[input_cols])
inputs_df[input_cols] = scaler.transform(inputs_df[input_cols])
```

###Create training and validation sets

```python
from sklearn.model_selection import train_test_split
train_inputs, val_inputs, train_targets, val_targets = train_test_split(
    inputs_df[input_cols], targets_df, test_size=0.25, random_state=22)
```
