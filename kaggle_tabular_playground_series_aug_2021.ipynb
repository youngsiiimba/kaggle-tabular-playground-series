```python
raw_train_df = pd.read_csv("./tabular-playground-series-aug-2021/train.csv")
raw_test_df = pd.read_csv("./tabular-playground-series-aug-2021/test.csv")
submission_df = pd.read_csv("./tabular-playground-series-aug-2021/sample_submission.csv")
```

## Clean data and prepare dataset for training

```python
nan_train_df = raw_train_df[raw_train_df.isna().any(axis=1)]
nan_test_df = raw_test_df[raw_test_df.isna().any(axis=1)]
```

```python
nan_train_df
```

```python
nan_test_df
```

###Identify input and target columns

```python
input_cols = raw_train_df.columns[1:-1]
target_cols = raw_train_df.columns[-1]
```

```python
input_cols
```

### Scale numeric values
 

1.   Let's scale the numeric values to a range of 0 to 1; Making it easier for our model to train
2.   In our case, all of our input columns are numeric values




```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(raw_train_df[input_cols])
```

```python
raw_train_df[input_cols] = scaler.transform(raw_train_df[input_cols])
raw_test_df[input_cols] = scaler.transform(raw_test_df[input_cols]) 
```

```python
raw_train_df
```

```python
raw_test_df
```

```python

test_inputs = raw_test_df[input_cols]
```

```python
from sklearn.model_selection import train_test_split

train_inputs, val_inputs, train_targets, val_targets = train_test_split(
    raw_train_df[input_cols], raw_train_df[target_cols], test_size=0.2, random_state=7)
```

```python
jovian.commit(project="kaggle-tabular-playground-series-aug-2021")
```

## Create and train models

### Evaluation
We'll use root mean squared error to evaluate our model

```python
from sklearn.metrics import mean_squared_error

def rmse(predictions, targets):
  return mean_squared_error(predictions, targets, squared=False)
```

### Gradient boosting machine

```python
import xgboost
from xgboost import XGBRegressor
```

```python
model1 = XGBRegressor(random_state=7, n_jobs=-1, n_estimators=20, max_depth=10)
```

```python
%%time
model1.fit(train_inputs, train_targets)
```

```python
print("Training RMSE: {}, Validation RMSE: {}".format(rmse(model1.predict(train_inputs), train_targets) ,rmse(model1.predict(val_inputs), val_targets)))
```

#### Experimenting with different hyperparameters

```python
def test_xgbregressor_params(**params):
  xgbmodel = XGBRegressor(random_state=7, n_jobs=-1, **params)
  xgbmodel.fit(train_inputs, train_targets)
  train_rmse = rmse(xgbmodel.predict(train_inputs), train_targets)
  val_rmse = rmse(xgbmodel.predict(val_inputs), val_targets)
  print('Train RMSE: {}, Validation RMSE: {}'.format(train_rmse, val_rmse))
```

```python
for i in range(70, 150, 10):
  print("Training with {} estimators".format(i))
  test_xgbregressor_params(n_estimators=i)
```

### LightGBM Regressor

```python
import numpy
from numpy import mean
from numpy import std
import lightgbm
from lightgbm import LGBMRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
```

```python
%%time
model2 = LGBMRegressor(random_state=7, n_jobs=-1)
# evaluate the model
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(model2, train_inputs, train_targets, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('MAE: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
model2.fit(train_inputs, train_targets)
```

```python
print("Training RMSE: {}, Validation RMSE: {}".format(rmse(model2.predict(train_inputs), train_targets), rmse(model2.predict(val_inputs), val_targets)))
```
